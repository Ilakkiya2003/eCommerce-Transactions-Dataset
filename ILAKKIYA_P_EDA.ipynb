```python
# 1. Correlation Analysis (Numerical Columns)
correlation_matrix = transactions[["Price", "Quantity", "TotalValue"]].corr()
print("Correlation Matrix:\n", correlation_matrix)

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Between Price, Quantity, and Total Value")
plt.show()
```

```python
# EDA - Visualizations
# 1. Customer Region Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=customers, x='Region', order=customers['Region'].value_counts().index)
plt.title("Customer Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Product Category Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=products, x='Category', order=products['Category'].value_counts().index)
plt.title("Product Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Transactions Over Time
transactions_by_date = transactions.groupby('TransactionDate').size()
plt.figure(figsize=(10, 5))
transactions_by_date.plot()
plt.title("Transactions Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
# 4. Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(transactions['Price'], kde=True, bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Total Transaction Value by Region
transactions_region = transactions.merge(customers, on="CustomerID")
transactions_region_grouped = transactions_region.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
transactions_region_grouped.plot(kind='bar')
plt.title("Total Transaction Value by Region")
plt.xlabel("Region")
plt.ylabel("Total Transaction Value")
plt.show()
```

```python
# 2. Seasonal trends in transactions
transactions["Month"] = transactions["TransactionDate"].dt.month
monthly_sales = transactions.groupby("Month")["TotalValue"].sum()

print("Monthly Sales Trends:\n", monthly_sales)

plt.figure(figsize=(8, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="o")
plt.title("Seasonal Trends in Transactions")
plt.xlabel("Month")
plt.ylabel("Total Transaction Value")
plt.xticks(range(1, 13))
plt.show()
```

```python
# 4. Average transaction value and its distribution
avg_transaction_value = transactions["TotalValue"].mean()
print(f"Average Transaction Value: ${avg_transaction_value:.2f}")

plt.figure(figsize=(8, 5))
sns.histplot(transactions["TotalValue"], bins=30, kde=True, color="skyblue")
plt.axvline(avg_transaction_value, color="red", linestyle="--", label="Average Value")
plt.title("Transaction Value Distribution")
plt.xlabel("Transaction Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Customer segments by total transaction value
customer_sales = transactions.groupby("CustomerID")["TotalValue"].sum().sort_values(ascending=False)
print("Top Customer Segments by Total Transaction Value:\n", customer_sales.head())

plt.figure(figsize=(8, 5))
sns.barplot(x=customer_sales.head().index, y=customer_sales.head().values, palette="Blues_d")
plt.title("Top Customer Segments by Total Transaction Value")
plt.xlabel("Customer ID")
plt.ylabel("Total Transaction Value")
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract day of the week and calculate total sales by day
transactions["DayOfWeek"] = transactions["TransactionDate"].dt.day_name()
sales_by_day = transactions.groupby("DayOfWeek")["TotalValue"].sum().sort_values(ascending=False)
print("Sales Trends by Day of the Week:\n", sales_by_day)

# Reordering days for a better visualization
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sales_by_day = sales_by_day[days_order]

plt.figure(figsize=(8, 5))
sns.barplot(x=sales_by_day.index, y=sales_by_day.values, palette="coolwarm")
plt.title("Sales Trends by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.show()
```
